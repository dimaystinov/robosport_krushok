```python
import cv2

video_path = "/Users/elijah/Documents/yolo/data_30_11/2019_11_29_21h49m11s/video.avi"

video = cv2.VideoCapture (video_path)

i = 0

start_ind = 0
stop_ind  = 1000000

while (True):
    ret, frame = video.read ()
    
    if (ret == True):
        print ("successfully read frame ", i)
    
    if (ret == False):
        print ("frame not read, exiting")
        break
    
    if (i > stop_ind):
        print ("writing finished, exiting")
        break
    
    cv2.imshow ("frame", frame)
    
    if (start_ind < i < stop_ind):
        print ("writing", i, "frame")
        cv2.imwrite ("/Users/elijah/Documents/yolo/data_30_11/2019_11_29_21h49m11s/images/frame" + str (i) + ".jpg", frame)
    
    cv2.waitKey (1)
    
    i += 1
    
```
