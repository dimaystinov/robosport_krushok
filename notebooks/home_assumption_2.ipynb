# These shoes were made for detecting

Основываясь на методах, рассказанных на втором семинаре, реализуйте детектор обуви на поле. Он не обязан работать идеально, достаточно того, что он дает разумные результаты. Обувь на поле - это тапки и ботинки, см. картинку ниже.

Источник (Source) поддерживает вход в виде папки с изображениями: метод get_frame() будет возвращать кадры в порядке сортировки их имен, если при инициализации подать ему путь до папки. Изображения находятся в data/images/log_video_frames. Воспроизведение закольцовано, то есть после последнего изображения get_frame() снова вернет первое.

Если с тем, чтобы реализовать все методами фреймворка, возникают технические проблемы, то можно или не пользоваться им в этих местах, или задать вопрос организаторам курса.

После выполнения задания в отдельной клетке после кода ответьте пожалуйста на следующие вопросы:

1) В каких случаях написанный Вами детектор работает не идеально?

2) Почему по Вашему мнению это происходит?

3) Какие методы обработки и анализа изображений могут помочь улучшить детекцию?

P.S. Обувь - это объект на зелёном, не круглый, но и не очень вытянутый, достаточно плотный и с ограничениями на размер в кадре сверху и снизу. Один из возможных методов решения - применить фильтр зеленого, сделать маску менее шумной, инвертировать ее (с помощью подачи в кастомный фильтр функции cv2.bitwise_not), найти связные компоненты, отвечающие обуви (отфильтровать), визуализировать.
