```python
import cv2
import time
import os
import math
import sys

sys.path.append("../modules/")

import input_output
import processor

det = processor.Processors ()
det.add_processor ("basket")

#det.add_filter (processor.gamma_correction (1), "basket", "gamma correction")
#cv2.namedWindow ('trackbars')
#cv2.createTrackbar ("gamma", "trackbars", 100, 200, 
#    lambda new_coeff : det.processors ["basket"] ["gamma correction"].set_gamma (float (new_coeff) / 100))

#det.add_filter (processor.crop (50, 50, 450, 200), "basket", "crop")

#low_th  = (25, 35, 80)
#high_th = (50, 55, 115)

low_th  = (25, 35, 80)
high_th = (50, 55, 115)

det.add_filter (processor.inrange (low_th, high_th), "basket", "inrange rgb")

#bbox extraction
det.add_filter (processor.max_area_cc_bbox (3), "basket", "bbox extraction")

source = input_output.Source ("../data/images/rgb_basket.png")

while (True):
    frame = source.get_frame ()

    mask, success = det.process (frame, "basket")

    stages = det.get_stages_picts ("basket")

    cv2.imshow ("rgb", input_output.form_grid (stages, window_x_sz=1000))
        
    time.sleep (0.02)

    keyb = cv2.waitKey (1) & 0xFF

    if (keyb == ord('q')):
        break

source.release()
cv2.destroyAllWindows()
```
